````python
INPUTS: 
Customer, Country, Amount, Risk Rating, Counterparty, Adverse Media, Network Link
````

````python
OUTPUTS: 
AI-style investigator enrichment summary
Risk drivers
Recommended next action
Draft case note
````

````python
## SAMPLE ALERT DATA:

import pandas as pd

alerts = [
    {
        "Customer": "Delta Commodities",
        "Country": "Russia",
        "Amount": 38000,
        "UsualAverage": 8000,
        "RiskRating": "High",
        "Counterparty": "Volga Trade",
        "NewCounterparty": "Yes",
        "AdverseMedia": "Yes",
        "NetworkLink": "Connected to high-risk trading network"
    },
    {
        "Customer": "Global Metals Ltd",
        "Country": "UAE",
        "Amount": 45000,
        "UsualAverage": 9000,
        "RiskRating": "Medium",
        "Counterparty": "Al Noor Trading",
        "NewCounterparty": "Yes",
        "AdverseMedia": "No",
        "NetworkLink": "No known network concern"
    }
]

df = pd.DataFrame(alerts)
df
````

````python
## SCORING LOGIC:

def calculate_risk(row):
    score = 0
    reasons = []

    if row["Country"] in ["Russia", "Iran", "North Korea", "UAE", "Turkey"]:
        score += 30
        reasons.append("High-risk jurisdiction exposure")

    if row["Amount"] > row["UsualAverage"] * 3:
        score += 25
        reasons.append("Transaction value significantly above usual behaviour")

    if row["NewCounterparty"] == "Yes":
        score += 20
        reasons.append("New counterparty relationship")

    if row["AdverseMedia"] == "Yes":
        score += 25
        reasons.append("Adverse media indicator")

    if row["RiskRating"] == "High":
        score += 15
        reasons.append("High customer risk rating")

    if "high-risk" in row["NetworkLink"].lower():
        score += 30
        reasons.append("Network relationship risk")

    return score, reasons
````

````python
## ENRICHMENT SUMMARY:

def investigator_enrichment(row):
    reasons = ", ".join(row["RiskReasons"])

    if row["RiskScore"] >= 100:
        recommendation = "Escalate for enhanced investigation and senior review."
    elif row["RiskScore"] >= 50:
        recommendation = "Prioritise for investigator review."
    else:
        recommendation = "Review under standard alert handling process."

    summary = f"""
Customer: {row['Customer']}
Counterparty: {row['Counterparty']}
Country: {row['Country']}
Risk Score: {row['RiskScore']}

Key Risk Drivers:
{reasons}

AI Investigator Enrichment:
The alert relates to {row['Customer']} and counterparty {row['Counterparty']}. 
The transaction amount of {row['Amount']} is being assessed against a usual average of {row['UsualAverage']}. 
Identified risk indicators include: {reasons}.

Recommended Action:
{recommendation}

Draft Case Note:
This alert has been enriched using customer, transaction, counterparty, adverse media and network relationship indicators. 
The case should be reviewed to determine whether the activity is consistent with the customer's expected profile and whether escalation is required.
"""
    return summary
````

````python
df[["RiskScore", "RiskReasons"]] = df.apply(
    lambda row: pd.Series(calculate_risk(row)),
    axis=1
)

df["InvestigatorEnrichment"] = df.apply(investigator_enrichment, axis=1)

df[["Customer", "RiskScore", "RiskReasons", "InvestigatorEnrichment"]]
````
